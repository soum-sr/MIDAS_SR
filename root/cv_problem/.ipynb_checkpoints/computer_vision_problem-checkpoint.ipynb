<h2>Data preprocessing</h2>
<br>
Importing numpy and pickle for data preprocessing and loading

```python
import numpy as np
import pickle 
```

Reading the pickle files and saving them to a variable

```python
with open('test_image.pkl', 'rb') as f:
    test_data = pickle.load(f)
with open('train_image.pkl', 'rb') as f:
    train_data = pickle.load(f)
with open('train_label.pkl', 'rb') as f:
    train_label = pickle.load(f)
```

Converting the saved variables to numpy arrays because the machine learning model only accepts arrays as inputs not lists

```python
test_data = np.array(test_data)
train_data = np.array(train_data)
train_label = np.array(train_label)
print("test data :", test_data.shape)
print("train data :", train_data.shape)
print("train label :", train_label.shape)
```

Reshaping the linear data to a 2-D array for visualization of data

```python
train_x = train_data.reshape(len(train_data), 28,28)
train_y = train_label
test_x = test_data.reshape(len(test_data), 28,28)
print(train_x.shape,train_y.shape, test_x.shape)
```

<b>Normalization of data.</b> <br>It is divided by 255 because the intensity of a color lies from (0,255) so dividing by 255 will make the range from (0,1)

```python
train_x = train_x /255.0
test_x = test_x / 255.0
```

<b>Visualization of data along with their labels</b>
<br><br>
Using matplotlib created a 5x5 subplot to plot 25 images along with their labels<br><br>
<i>Here one thing to notice that the given train_label data and the given train_image data's indexs are not matching, but still continued with the data provided</i>

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize = (10, 10))
for i in range(25):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid('off')
    plt.imshow(train_x[i], cmap=plt.cm.binary)
    plt.xlabel(train_label[i])
```

<h2>Deep Learning Model</h2>

Importing keras and tensorflow for layers and activation functions

```python
import keras
import tensorflow as tf
```

<b>FLATTEN</b>: <i>InputLayer</i> converts the 2D array to a 1D array<br>
<b>Dense(128)</b>: <i>HiddenLayer</i> contains 128 neurons <br>
<b>Dense(7)</b>:<i>OutputLayer</i> it has 7 layers because the train_label has max_label = 6 so it predicts for 7 classes from 0 to 6, since it takes the indexes as classes

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(7,activation=tf.nn.softmax)
                         
])
```

Compiling the model with <i>Optimizer</i> as <b>Adam Optimizer</b> and <i>Loss function</i> as <b>Categorical Crossentropy</b>

```python
model.compile(optimizer=tf.train.AdamOptimizer(), 
             loss = 'sparse_categorical_crossentropy',
             metrics=['accuracy'])
```

Fitting the training data to model and got about 90% accuracy<br>
<i>One thing to notice that the accuracy might increase if the train_data is labeled correctly</i>

```python
model.fit(train_x, train_label, epochs = 20)
```

Getting the Predictions on test data

```python
predictions = model.predict(test_x)
```

Output array predicted by the model from the test data

```python
predictions
```

<h2>Saving the Predictions</h2

```python
out_data = []
for i, pred in enumerate(predictions):
    x = []
    x.append(i)
    x.append(np.argmax(pred))
    out_data.append(x)

np.savetxt("Soumyajit_Rout.csv", out_data,delimiter = ",")
```
