<h3>Importing Required libraries</h3>

```python
from tweepy import OAuthHandler
from tweepy import API
from tweepy import Cursor
from datetime import datetime, date, time , timedelta
from collections import Counter
import sys
```

<h3>Secret keys</h3>

```python
consumer_key=""
consumer_secret=""
access_token=""
access_token_secret=""
```

<h3>Authorizing the secret keys and accessing the tweeter api</h3>

```python
auth = OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
auth_api = API(auth)
```

<h3>Getting the tweets of an user</h3>

```python
user = 'midasIIITD'
```

```python
tweets = auth_api.user_timeline(screen_name=user)
tweets
```

<h3>The required attributes</h3>

```python
tweet_text = []
tweet_time = []
tweet_fav_count = []
tweet_retweet_count = []
tweet_img_count = []
final = []
```

<h3>Going through all tweets and appending them in a dict and appending the dict to a list</h3>

```python
for index, tweet in enumerate(tweets):
    tweet_text.append(tweet.text)
    tweet_time.append(tweet.created_at)
    tweet_retweet_count.append(tweet.retweet_count)
    tweet_fav_count.append(tweet.favorite_count)
    media = tweet.entities.get('media', [])
    count = 0 
    for med in media:
        count += 1
    if count == 0:
        tweet_img_count.append("NULL")
    else:
        tweet_img_count.append(count)
    output = {
    "text":tweet_text[index],
    "time":tweet_time[index],
    "fav_count":tweet_retweet_count[index],
    "retweet_count":tweet_retweet_count[index],
    "img_count":tweet_img_count[index]
    }
    final.append(output)
```

```python
final
```

<h3>Dumping the data to a json file</h3>

```python
import json
with open('result.json', 'w') as f:
    f.write(json.dumps(final,default=str))
```
